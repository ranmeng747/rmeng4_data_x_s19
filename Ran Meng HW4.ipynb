#### Q 4.
The data files needed: 

`google_data.txt, ny_temps.txt & yahoo_data.txt`

Use your knowledge with `Python, NumPy, pandas and matplotlib` to reproduce the plot below:

![](answer.png?raw=true)

```python
# Importing NY data
ny = pd.read_csv('ny_temps.txt', sep="\t")
ny_date = ny.iloc[:, 0]
ny_temp = ny.iloc[:, 1]

# Importing Google data
google = pd.read_csv('google_data.txt', sep="\t")
google_date = google.iloc[:, 0]
google_value = google.iloc[:, 1]

# Importing Yahoo data
yahoo = pd.read_csv('yahoo_data.txt', sep="\t")
yahoo_date = yahoo.iloc[:, 0]
yahoo_value = yahoo.iloc[:, 1]
```

```python
f, ax = plt.subplots(figsize = (12, 8))

l_google, = ax.plot(google_date, google_value, color = 'green', label = "Google Stock Value")
l_yahoo, = ax.plot(yahoo_date, yahoo_value, color = 'purple', label = "Yahoo! Stock Value")
ax.set_xlim(48800, 55600)
ax.set_xlabel("Date(MJD)", fontsize = 16, labelpad = 20 )
ax.set_ylabel('Value(Dollars)', color = 'purple', fontsize = 14)
ax.tick_params('y',colors='green')
ax.set_title("New York Temperature, Google, and Yahoo!", fontsize = 20, fontweight = 'bold', pad = 40)

#ax2 for NY temp
ax2 = ax.twinx()
l_ny, = ax2.plot(ny_date, ny_temp, linestyle = "--", color = 'blue', label = "NY Mon. High Temp")
ax2.set_ylim(-150, 100)
ax2.set_ylabel('Temperature (F)', color = 'blue', fontsize = 14)
ax2.tick_params('y',colors='blue')



#Adding minor tickets
x_ticks = []
for i in range (49000, 55600, 200):
    x_ticks.append(i)

y_ticks = []
for i in range(0, 760, 20):
    y_ticks.append(i)

y2_ticks = []
for i in range(-150, 100, 10):
    y2_ticks.append(i);

ax.set_xticks(x_ticks, minor = True)    
ax.set_yticks(y_ticks, minor = True)
ax2.set_yticks(y2_ticks, minor = True)

# Add legend
plt.legend((l_yahoo,l_google, l_ny),("Yahoo! Stock Value","Google Stock Value", "NY Mon. High Temp"),loc=[0.02,0.45],
           fontsize = 12, frameon = False)

plt.show()
```
